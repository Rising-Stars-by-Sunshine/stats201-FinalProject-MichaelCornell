```python
import pandas as pd
import datetime
```

```python
natData = pd.read_csv("Trips_by_Distance (1).csv")
natData.head()
```

```python
natData = natData.dropna(subset=['County FIPS'])
natData.head()
```

```python
natData['Date'] = pd.to_datetime(natData['Date'])

res = natData[~(natData['Date'] < '2022-01-01')]

print(res.Week.unique())
```

```python
colNames = []
for col in res.columns:
    colNames.append(col)
print(colNames)
```

```python
avg = res.groupby(['County FIPS'])[colNames].mean()
avg
```

```python
res = res.drop(columns=['Date', 'State FIPS', 'Population Staying at Home', 'Population Not Staying at Home', 'Number of Trips', 'Number of Trips <1', 'Number of Trips 1-3', 'Number of Trips 3-5', 'Number of Trips 5-10', 'Number of Trips 10-25', 'Number of Trips 25-50', 'Number of Trips 50-100', 'Number of Trips 100-250', 'Number of Trips 250-500', 'Number of Trips >=500', 'Row ID', 'Week', 'Month'])
```

```python
res = res.drop_duplicates(ignore_index=True)
res
```

```python
res = res.sort_values(by = 'County FIPS')
res = res.reset_index()
res = res.drop(columns=['index'])
res.sort_index()
avg = avg.sort_index(level='County FIPS')
avg = avg.reset_index(drop=True)
avg.sort_index()
```

```python
merged_data = avg.join(res, how='left', lsuffix='_l', rsuffix='_r')
merged_data = merged_data.drop(columns=['County FIPS_r']).rename(columns={"County FIPS_l": "County FIPS"})
```

```python
merged_data
```

```python
walkScore = pd.read_csv("EPA_SmartLocationDatabase_V3_Jan_2021_Final.csv")
walkScore['COUNTYFP'] = walkScore['STATEFP']*1000 + walkScore['COUNTYFP']
walkScore.head()
```

```python
colNamesWalk = []
for col in walkScore.columns:
    colNamesWalk.append(col)
print(colNamesWalk)
```

```python
walkScore['totWalkInd'] = walkScore['NatWalkInd'] * walkScore['TotPop']
```

```python
countySum = walkScore.groupby(['COUNTYFP'])['Ac_Total', 'Ac_Water', 'Ac_Land', 'Ac_Unpr', 'TotPop', 'CountHU', 'HH', 'AutoOwn0', 'AutoOwn1', 'AutoOwn2p', 'Workers', 'R_LowWageWk', 'R_MedWageWk', 'R_HiWageWk', 'TotEmp', 'E5_Ret', 'E5_Off', 'E5_Ind', 'E5_Svc', 'E5_Ent', 'E8_Ret', 'E8_off', 'E8_Ind', 'E8_Svc', 'E8_Ent', 'E8_Ed', 'E8_Hlth', 'E8_Pub', 'E_LowWageWk', 'E_MedWageWk', 'E_HiWageWk', 'D1A', 'D1B', 'D1C', 'D1C5_RET', 'D1C5_OFF', 'D1C5_IND', 'D1C5_SVC', 'D1C5_ENT', 'D1C8_RET', 'D1C8_OFF', 'D1C8_IND', 'D1C8_SVC', 'D1C8_ENT', 'D1C8_ED', 'D1C8_HLTH', 'D1C8_PUB', 'D1D', 'D1_FLAG', 'D2A_JPHH', 'D2B_E5MIX', 'D2B_E5MIXA', 'D2B_E8MIX', 'D2B_E8MIXA', 'D2A_EPHHM', 'D2C_TRPMX1', 'D2C_TRPMX2', 'D2C_TRIPEQ', 'D2R_JOBPOP', 'D2R_WRKEMP', 'D2A_WRKEMP', 'D2C_WREMLX', 'D3A', 'D3AAO', 'D3AMM', 'D3APO', 'D3B', 'D3BAO', 'D3BMM3', 'D3BMM4', 'D3BPO3', 'D3BPO4', 'D4A', 'D4B025', 'D4B050', 'D4C', 'D4D', 'D4E', 'D5AR', 'D5AE', 'D5BR', 'D5BE', 'D5CR', 'D5CRI', 'D5CE', 'D5CEI', 'D5DR', 'D5DRI', 'D5DE', 'D5DEI', 'Shape_Area', 'totWalkInd'].sum()
countySum = countySum.sort_values(by = 'COUNTYFP')
countySum = countySum.reset_index().rename(columns={"COUNTYFP": "County FIPS"})
countySum.drop(countySum[countySum['County FIPS'] >= 56046].index, inplace = True)
countySum['weightedIndex'] = countySum['totWalkInd'] / countySum['TotPop']
countySum
```

```python
data_full = merged_data.join(countySum, how='left', lsuffix='_l', rsuffix='_r')
data_full = data_full.drop(columns=['County FIPS_r', 'Week', 'Month']).rename(columns={"County FIPS_l": "County FIPS"})
data_full
```

```python
data_full.to_csv("TotalWalkDrivingData.csv")
```

```python
causal_data = data_full.drop(columns=['Ac_Total', 'Ac_Water', 'CountHU', 'HH', 'E5_Ret', 'E5_Off', 'E5_Ind', 'E5_Svc', 'E5_Ent', 'E8_Ret', 'E8_off', 'E8_Ind', 'E8_Svc', 'E8_Ent', 'E8_Ed', 'E8_Hlth', 'E8_Pub', 'E_LowWageWk', 'E_MedWageWk', 'E_HiWageWk', 'D1A', 'D1B', 'D1C', 'D1C5_RET', 'D1C5_OFF', 'D1C5_IND', 'D1C5_SVC', 'D1C5_ENT', 'D1C8_RET', 'D1C8_OFF', 'D1C8_IND', 'D1C8_SVC', 'D1C8_ENT', 'D1C8_ED', 'D1C8_HLTH', 'D1C8_PUB', 'D1D', 'D1_FLAG', 'D2A_JPHH', 'D2B_E5MIX', 'D2B_E5MIXA', 'D2B_E8MIX', 'D2B_E8MIXA', 'D2A_EPHHM', 'D2C_TRPMX1', 'D2C_TRPMX2', 'D2C_TRIPEQ', 'D2R_JOBPOP', 'D2R_WRKEMP', 'D2A_WRKEMP', 'D2C_WREMLX', 'D3A', 'D3AAO', 'D3AMM', 'D3APO', 'D3B', 'D3BAO', 'D3BMM3', 'D3BMM4', 'D3BPO3', 'D3BPO4', 'D4A', 'D4B025', 'D4B050', 'D4C', 'D4D', 'D4E', 'D5AR', 'D5AE', 'D5BR', 'D5BE', 'D5CR', 'D5CRI', 'D5CE', 'D5CEI', 'D5DR', 'D5DRI', 'D5DE', 'D5DEI', 'Shape_Area', 'TotEmp'])
causal_data
```

```python
causal_data_colNames = []
for col in causal_data.columns:
    causal_data_colNames.append(col)
for x in ['State FIPS', 'County FIPS', 'Level', 'State Postal Code', 'County Name', 'totWalkInd', 'weightedIndex', 'TotPop', 'R_LowWageWk', 'R_MedWageWk', 'R_HiWageWk']:
    causal_data_colNames.remove(x)
print(causal_data_colNames)
```

```python
for name in causal_data_colNames:
    if name in ['Population Staying at Home', 'Population Not Staying at Home', 'AutoOwn0', 'AutoOwn1', 'AutoOwn2p', 'Workers']:
        causal_data['percent_' + name] = causal_data[name] / causal_data['TotPop']
    else:
        causal_data[name + '_perPerson'] = causal_data[name] / causal_data['TotPop']
for name in ['R_LowWageWk', 'R_MedWageWk', 'R_HiWageWk']: #'Workers'
    causal_data['percent_' + name] = causal_data[name] / causal_data['Workers']
causal_data
```
